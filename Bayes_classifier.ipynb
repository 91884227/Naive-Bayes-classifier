```python
def num_pixel_pro(pixser_):
    #done_pix = pixser_.tolist()
    done_pix = list(pixser_)
    buf = []
    for i in np.arange(32):
        for_buf = done_pix.count(i)
        if( for_buf == 0):
            buf = buf + [1]
        else:
            buf = buf + [for_buf]
    
    buf = np.array(buf)/sum(buf)
    return(buf)
```

```python
def Bayes_classifier(pixelnp_):
    buf = []
    for i in np.arange(10):
        buf = buf + [num_pix_likli2(pixelnp_, i)]
    buf = np.array(buf)
    buf = buf / buf.sum()
    return(buf)
```
