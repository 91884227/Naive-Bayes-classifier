```python
import numpy as np
import pandas as pd
import math
from scipy.stats import norm
pd.set_option('display.max_rows', 30)
pd.set_option('display.max_columns', 30)
pd.set_option('display.width', 1000)
```

```python
# need to add variance
def number_par_set(int_):
    step_df = train_df[train_df["label"] == int_ ]
    print(int_)
    done_df = step_df.drop(['label'], axis = 1)
    buf = done_df.describe().T[["mean", "std"]]
    buf["std2"] =  buf[buf["std"]>0]["std"].min()
    buf["std3"] = buf[["std", "std2"]].T.max().T
    return( buf)
```

```python
def num_pix_likli2(pix_, num_):
    buf = 0
    num_df = number_pro_pd[num_]
    for i, x in enumerate(pix_):
        mean = num_df.iloc[i, 0]
        var = num_df.iloc[i, 3]
        #print(norm.pdf(x, loc= mean, scale=var))
        try:
            #print(math.log(norm.pdf(x, loc= mean, scale=var)))
            buf = buf + math.log(norm.pdf(x, loc= mean, scale=var)) 
        except:
            buf = buf - 400
    return(buf)
```

```python
def Gaussian_classifier(pixelnp_):
    buf = []
    for i in np.arange(10):
        buf = buf + [num_pix_likli2(pixelnp_, i)]
    buf = np.array(buf)
    buf = buf / buf.sum()
    return(buf)
```

```python
def Gaussian_outcome(index_):
    prob = Gaussian_classifier(test_X.to_numpy()[index_])
    print("Postirior (in log scale):")
    for i in np.arange(10):
        print("%d: %.4f" % (i, prob[i]))
    print( "Predicrtion: %d , Ans: %d\n"% (np.argmin(prob), test_y[index_]))
    global error_count_glo
    if( np.argmin(prob) != test_y[index_]):
        error_count_glo = error_count_glo + 1
```

```python
def print_image(int_):
    buf = np.array(number_pro_pd[int_]["mean"]>=128).astype(int)
    print(pd.DataFrame(buf.reshape((28, 28))))
```
